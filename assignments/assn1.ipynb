## Problem 1 [20 Points]

### Subproblem 1 [10 Points]

Consider a coin that comes up heads with probability $p$ and tails with probability $1-p$. Let $q_n$ be the probability of obtaining even number of heads in $n$ independent tosses. Derive a recursion that relates $q_n$ to $q_{n-1}$ and establish the formula
$$q_n = \frac{1+(1-2p)^n}{2}$$


### Subproblem 2 [10 Points]

Let $X$ and $Y$ have joint PDF 
$$
f_{X,Y}(x,y) = \begin{cases}
C e^{-(ax+by)} & x,y \geq 0\\
0 & \text{otherwise}
\end{cases},
$$

where, $a, b > 0$ are constants

- Determine the constant $C$
- Find the marginal density of $X$ and $Y$. What can you infer from them?
- Find $\mathbb{E}(Y \mid X> \frac{\exp(a^2 + b^2)}{a^4 + b^4})$


## Answer:

### Subproblem 1

We condition on the $n^{\text{th}}$ toss

\begin{align}
q_n &= \mathbb{P}(\text{Obtaining even number of heads in $n$ independent tosses}|\text{Last toss is a head})\mathbb{P}(\text{Last toss is a head}) + \mathbb{P}(\text{Obtaining even number of heads in $n$ independent tosses}|\text{Last toss is a tail})\mathbb{P}(\text{Last toss is a tail}) \\
q_n &= \mathbb{P}(\text{Obtaining odd number of heads in $n - 1$ independent tosses})\mathbb{P}(\text{Last toss is a head}) + \mathbb{P}(\text{Obtaining even number of heads in $n-1$ independent tosses})\mathbb{P}(\text{Last toss is a tail}) \\
q_n &= (1 - q_{n-1})p + q_{n-1}(1-p) \\
&= p + (1 - 2p)q_{n-1}
\end{align}

Note that $q_1 = 1 - p$. Hence we have that
\begin{align}
q_n &= p + (1 - 2p)(p + (1-2p)q_{n-2}) \\
    &= p + p(1-2p) + (1-2p)^2 q_{n-2} \\
    &= p + p(1-2p) + (1-2p)^2 (p + (1-2p)q_{n-3}) \\
    &= p + p(1-2p) + p(1-2p)^2 + (1-2p)^3q_{n-3} \\
    &\vdots \\
    &= p(1 + (1-2p) + (1-2p)^2 + \cdots + (1 - 2p)^{n-2}) + (1-2p)^{n-1}q_1 \\
    &= p(1 + (1-2p) + (1-2p)^2 + \cdots + (1 - 2p)^{n-2}) + (1-2p)^{n-1}q_1 \\
    &= p\frac{(1 - (1-2p)^{n-1})}{1 - (1-2p)} + (1-2p)^{n-1} (1 - p) \\
    &= p\frac{(1 - (1-2p)^{n-1})}{2p} + (1-2p)^{n-1} (1 - p) \\
    &= \frac{(1 - (1-2p)^{n-1})}{2} + (1-2p)^{n-1} (1 - p) \\
    &= \frac{(1 - (1-2p)^{n-1})}{2} + \frac{(1-2p)^{n-1} (2 - 2p)}{2} \\
    &= \frac{(1 - (1-2p)^{n-1} + (1-2p)^{n-1} (2 - 2p))}{2} \\
    &= \frac{(1 + (1-2p)^{n-1} - 2p(1-2p)^{n-1})}{2} \\
    &= \frac{(1 + (1-2p)^{n-1}(1 - 2p))}{2} \\
    &= \frac{(1 + (1-2p)^{n})}{2}
\end{align}

### Subproblem 2

- Using that $\int_{\mathbb{R}^2} f_{X, Y}(x, y) = 1$, we have $C = ab$
- We have

\begin{align}
    f_X(x) = \int_{-\infty}^{\infty} f_{X, Y} (x, y) dy = a e^{-ax} \\
    f_Y(y) = \int_{-\infty}^{\infty} f_{X, Y} (x, y) dx = b e^{-by}
\end{align}

We have that $f_{X, Y} (x, y) = f_X \cdot f_Y$. Hence, $X$ and $Y$ are independent.
- Since $X$ and $Y$ are independent, conditioning on $X$ doesn't affect $Y$. Hence,

\begin{align}
    \mathbb{E}(Y \mid X> \frac{\exp(a^2 + b^2)}{a^4 + b^4}) = \mathbb{E}(Y) = \int_0^{\infty} b y e^{-by} dy = \frac{1}{b}
\end{align}

## Problem 2 [20 points]

Let $S_1,S_2,\ldots,S_n$ be a partition of the sample space $\Omega$.
- Show that for any event $A$,
  $$\mathbb{P}(A) = \sum_{i=1}^n \mathbb{P}\left({A \cap S_i}\right)$$
- Use the previous part to show that, for events $A$, $B$ and $C$,
  $$\mathbb{P}\left({A}\right) = \mathbb{P}\left({A \cap B}\right) + \mathbb{P}\left({A \cap C}\right) + \mathbb{P}\left({A \cap B^c \cap C^c}\right) - \mathbb{P}\left({A \cap B \cap C}\right)$$
- Prove that for any two events $A$ and $B$, we have
$$\mathbb{P} (A \cap B) \geq \mathbb{P}(A) + \mathbb{P}(B) - 1$$
- Using the above, establish the following generalization:
$$\mathbb{P}(A_1 \cap A_2 \cap \cdots \cap A_n) \geq \mathbb{P}(A_1) + \mathbb{P}(A_2) + \cdots + \mathbb{P}(A_n) - (n-1)$$

**[You will need to argue logically using only the axioms. Drawing a diagram is not enough.]**

## Answer:

- Since $S_1, S_2, \cdots, S_n$ are partitions of $\Omega$, we have that $\Omega = \cup_{i = 1}^n S_i$ and $S_i \cap S_j = \emptyset$ for $i \neq j$. Hence, we have that
\begin{align}
A = A \cap \Omega &= A \cap \left(\cup_{i = 1}^n S_i\right) = \cup_{i = 1}^n \left(A \cap S_i\right) \\
\end{align}

Since $S_i$'s are mutually disjoint, so are $A \cap S_i$'s. Hence, we have that
\begin{align}
\mathbb{P}(A) = \mathbb{P}\left(\cup_{i = 1}^n \left(A \cap S_i\right)\right) = \sum_{i = 1}^n \mathbb{P}\left(A \cap S_i\right)
\end{align}

- Building upon the previous part, our sample space can be partitioned into disjoint set $S_1 = B \cup C$ and $S_2 = (B \cup C)^c = B^c \cap C^c$. Hence, we have that
\begin{align}
    \mathbb{P}\left(A\right) & = \mathbb{P}\left(A \cap S_1\right) + \mathbb{P}\left(A \cap S_2\right) = \mathbb{P}\left(A \cap \left(B \cup C\right)\right) + \mathbb{P} \left(A \cap \left(B^c \cap C^c\right)\right)\\
    & = \mathbb{P}\left(\left(A \cap B\right) \cup \left(A \cap C\right)\right) + \mathbb{P} \left(A \cap B^c \cap C^c\right)\\
    & = \mathbb{P}\left(A \cap B\right) + \mathbb{P}\left(A \cap C\right) - \mathbb{P}\left(\left(A \cap B\right) \cap \left(A \cap C\right)\right) + \mathbb{P} \left(A \cap B^c \cap C^c\right)\\
    & = \mathbb{P}\left(A \cap B\right) + \mathbb{P}\left(A \cap C\right) + \mathbb{P}\left(A \cap B^c \cap C^c\right) - \mathbb{P}\left(A \cap B \cap C\right)
\end{align}

- We know that $\mathbb{P}(A \cap B) = \mathbb{P}(A) + \mathbb{P}(B) - \mathbb{P}(A \cup B)$. Also, $\mathbb{P}(A \cup B) \leq 1$. Hence, we have that $\mathbb{P}(A \cap B) \geq \mathbb{P}(A) + \mathbb{P}(B) - 1$.

- Let $S_{n + 1} = \left(A_1 \cap A_2 \cap \cdots \cap A_n\right)$. Building upon the earlier result, we have that
\begin{align}
\mathbb{P}(S_{n + 1}) = \mathbb{P}(S_n \cap A_n) \geq \mathbb{P}(S_n) + \mathbb{P}(A_n) - 1
\end{align}
Proceeding inductively, we can show that
\begin{align}
\mathbb{P}(S_{n}) & \geq \sum_{i = 1}^{n - 1}\mathbb{P}(A_i) - (n - 2) \\
\implies \mathbb{P}(S_{n + 1}) & \geq \sum_{i = 1}^{n}\mathbb{P}(A_i) - (n - 1)
\end{align}

## Problem 4 [20 Points]

As mentioned in class the Gaussian has nice properties which makes it a fundamental tool in statistical infererence. The standard normal $\mathcal{N}(\mathbf{x}; \mathbf{\mu}, \mathbf{\Sigma})$ is defined as

$$
    \mathcal{N}(\mathbf{x}; \mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{\sqrt{(2 \pi)^n |\mathbf{\Sigma}|}} \exp \left[ -\frac{1}{2} (\mathbf{x} - \mathbf{\mu})^\top \Sigma^{-1} (\mathbf{x} - \mathbf{\mu}) \right]
$$

- Prove that if $x \in \mathbb{R}^d$ is normally distributed, every affine transformation $y = A x + b$ also has a Gaussian distribution. Find its mean and covariance.
- Analytically find the KL divergence $\mathbb{KL}(P || Q)$ between two multivariate normal distributions $p(x) \sim \mathcal{N}(\mathbf{x}; \mathbf{\mu}_1, \mathbf{\Sigma}_1)$ and $q(x) \sim \mathcal{N}(\mathbf{x}; \mathbf{\mu}_2, \mathbf{\Sigma}_2)$.

## Answer:

- If $y = A x + b$, we have that
\begin{align}
    \bar{y} = \mathbb{E}[y] &= \mathbb{E}[A x + b] \\
    &= \mathbb{E}[A x] + \mathbb{E}[b] \\
    &= A \mathbb{E}[x] + b \\
    &= A \mu + b
\end{align}
For the covariance, we have that 
\begin{align}
\Sigma_y = \mathbb{E}[(y - \bar{y})(y - \bar{y})^\top] &= \mathbb{E}[(A x + b - A \mu - b)(A x + b - A \mu - b)^\top] \\
&= \mathbb{E}[(A x - A \mu)(A x - A \mu)^\top] \\
&= A \mathbb{E}[(x - \mu)(x - \mu)^\top] A^\top\\
&= A \Sigma A^\top
\end{align}
Then, we can define
\begin{align}
    y \sim \mathcal{N}(\bar{y}, \Sigma_y)
\end{align}

- 

\begin{align}
    \mathbb{KL}(p||q) & = \int_{-\infty}^{\infty} p(x) \cdot \ln\left(\dfrac{p(x)}{q(x)}\right)dx\\
    & = \mathbb{E}_p \left[\ln p(x)  -\ln q(x) \right]dx\\
    & = \dfrac{1}{2} \mathbb{E}_p \left[ \ln \left( \dfrac{|\Sigma_q|}{|\Sigma_p|} \right) - \dfrac{1}{2} (\mathbf{x} -\mathbf{\mu}_p)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_p) + \dfrac{1}{2} (\mathbf{x} -\mathbf{\mu}_q)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_q)\right] \\
    & = \dfrac{1}{2} \ln \left( \dfrac{|\Sigma_q|}{|\Sigma_p|} \right) - \dfrac{1}{2} \mathbb{E}_p \left[ (\mathbf{x} -\mathbf{\mu}_p)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_p)\right] + \dfrac{1}{2} \mathbb{E}_p \left[ (\mathbf{x} -\mathbf{\mu}_q)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_q)\right] \\
\end{align}
Note that the second term above is a scalar. Hence, we can utilize the trace trick to simplify the expression as

\begin{align}
    \mathbb{E}_p \left[(\mathbf{x} -\mathbf{\mu}_p)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_p) \right] 
    & = \mathbb{E}_p \left[ \text{tr} \left[ (\mathbf{x} -\mathbf{\mu}_p)^\top \Sigma_p^{-1} (\mathbf{x} - \mathbf{\mu}_p)\right]\right] \\
    & = \text{tr} \left[ \mathbb{E}_p\left[ (\mathbf{x} -\mathbf{\mu}_p) (\mathbf{x} - \mathbf{\mu}_p)^\top \right]\Sigma_p^{-1} \right]\\
    & = \text{tr} \left[ \Sigma_p \Sigma_p^{-1} \right]\\
    & = \text{tr} \left[ I \right]\\
    & = d
\end{align}
where $\mathbf{x} \in \mathbb{R}^d$. 
Similarly,
\begin{align}
    \mathbb{E}_p \left[(\mathbf{x} -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{x} - \mathbf{\mu}_q) \right]
    & = (\mathbf{\mu}_p -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{\mu}_p - \mathbf{\mu}_q) + \text{tr} \left[\mathbb{E}_p\left[(\mathbf{x} -\mathbf{\mu}_p)  (\mathbf{x} - \mathbf{\mu}_p)^\top\Sigma_q^{-1}\right]\right]\\
    & = (\mathbf{\mu}_p -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{\mu}_p - \mathbf{\mu}_q) + \text{tr} \left[\mathbb{E}_p\left[(\mathbf{x} -\mathbf{\mu}_p)  (\mathbf{x} - \mathbf{\mu}_p)^\top\right]\Sigma_q^{-1}\right]\\
    & = (\mathbf{\mu}_p -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{\mu}_p - \mathbf{\mu}_q) + \text{tr} \left[ \Sigma_p\Sigma_q^{-1}\right]\\
\end{align}

$$ 
\implies \mathbb{KL}(p||q) = \dfrac{1}{2} \left[ \ln \left( \dfrac{|\Sigma_q|}{|\Sigma_p|} \right) - d +  (\mathbf{\mu}_p -\mathbf{\mu}_q)^\top \Sigma_q^{-1} (\mathbf{\mu}_p - \mathbf{\mu}_q) + \text{tr} \left[ \Sigma_p\Sigma_q^{-1}\right]\right]
$$

## Problem 5 [20 Points]

Consider a two-dimensional random variable $z=(z_1,z_2)$ distributed as

$$
p(z_1,z_2) = \mathcal{N}\left(\left[\begin{matrix} 0 \\ 0 \end{matrix}\right],
\left[\begin{matrix} 1 & 0 \\ 0 & 1 \end{matrix}\right]\right).
$$


Also consider a transformation $z_1 = g_1(x_1) = x_1^3 - 6x_1^2 + 12x_1 - 8$ whose inverse mapping is $x_1 = g_1^{-1}(z_1) = z_1^{1/3} + 2$, and $z_2 = g_2(x_2) = x_2 + 1$ whose inverse mapping is $x_2 = g_2^{-1}(z_2) = z_2 - 1$. Generate $5000$ realizations for $(z_1, z_2)$ from $p(z_1,z_2)$ and compute the corresponding $(x_1, x_2)$ pairs according to the given transformation. Then, maximize the log-likelihood of the $(x_1, x_2)$ observations to estimate the parameters of the forward mapping that transforms $p(z)$ into $p(x)$ (consider a polynomial basis). Report the learned parameters, log-likelihood loss and a visualization of $p(x)$ at every step of your gradient ascent algorithm.


```python
# importing the needed libraries
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '7'
import jax.numpy as np
from jax import random, jit, vmap, grad, jacfwd, tree_map
from jax.scipy.stats import multivariate_normal
from tqdm.auto import trange
import pylab as pl
from IPython.display import clear_output
```

```python
# Creating realizations for Z:
num_samples = 2000
rng_key     = random.PRNGKey(0)
Z           = random.normal(rng_key, (num_samples, 2))

# Defining the functions g and g_inv:
def g_inv(z):
    z1 = z[0]
    z2 = z[1]
    x1 = np.sign(z1) * abs(z1)**(1 / 3) + 2
    x2 = z2 - 1  
    return np.array([x1, x2])

def g(x, coeff_x1_basis, coeff_x2_basis):
    x1 = x[0]
    x2 = x[1]
    z1 = np.sum(x1**(np.arange(0, coeff_x1_basis.size)) * coeff_x1_basis)
    z2 = np.sum(x2**(np.arange(0, coeff_x2_basis.size)) * coeff_x2_basis)
    return np.array([z1, z2])

mu     = np.zeros(2)
Sigma  = np.eye(2)
logp_z = multivariate_normal.logpdf(Z, mean = mu, cov = Sigma)
```

```python
# Getting vmapped versions for g and g_inv:
g_inv_vec = vmap(g_inv, in_axes = (0,))
g_vec     = vmap(g, in_axes = (0, None, None))

# Generating pairs of X:
X = g_inv_vec(Z)
# Checking that the given forward and inverse transforms are consistent:
coeff1_true = np.array([-8, 12, -6, 1])
coeff2_true = np.array([1, 1])
assert(np.mean(abs(g_vec(X, coeff1_true, coeff2_true) - Z)) < 1e-6)
```

```python
J       = jacfwd(g)
# Getting vmapped versions of J and det:
J_vec   = vmap(J, in_axes = (0, None, None))
det_vec = vmap(np.linalg.det)
```

```python
# Framing the gradient ascent problem as a gradient descent:
@jit
def loss(params, batch):
    coeff1, coeff2 = params
    return -np.mean(multivariate_normal.logpdf(g_vec(batch, coeff1, coeff2), 
                                               mean = mu, cov = Sigma) + 
                    np.log(np.abs(det_vec(J_vec(batch, coeff1, coeff2)))))
    
# Computing Gradients:
grad_loss = jit(grad(loss))
```

```python
# Running this block will show a visualization of progress for p(x)
# We assume polynomial basis P_3 for z1:
coeff1_est = 4 * random.normal(rng_key, (4,))
# We assume polynomial basis P_1 for z1:
coeff2_est = random.normal(rng_key, (2,))

params = coeff1_est, coeff2_est
batch  = X

# Gradient descent
# Learning Rate
alpha = 0.0005
n_its = 500000

# Variables for Plotting:
x    = np.linspace(-5.0, 5.0, 401)
y    = np.linspace(-5.0, 5.0, 401)
X, Y = np.meshgrid(x, y)
grid = np.hstack([X.reshape(-1, 1), Y.reshape(-1, 1)])

loss_history = [loss(params, batch)]
frames = []
for i in trange(n_its):
    g = grad_loss(params, batch)
    params = tree_map(lambda x, g: x - alpha * g, params, g)

    # Track progress:
    loss_history = loss_history + [loss(params, batch)]

    # Uncomment to plot the progress:
    # if i % 1000 == 0:
    #     print('Iteration:', i)
    #     print('Loss:', loss(params, batch))

    #     coeff1_est, coeff2_est = params
    #     Z = np.exp(  multivariate_normal.logpdf(g_vec(grid, coeff1_est, coeff2_est), 
    #                                             mean = mu, cov = Sigma) 
    #                 + np.log(np.abs(det_vec(J_vec(grid, coeff1_est, coeff1_est)))))

    #     Z = Z.reshape(X.shape)

    #     clear_output(wait=True)
    #     pl.figure(dpi = 150)
    #     pl.title("Iteration = {}, Loss = {}".format(i, loss(params, batch)))
    #     pl.hist2d(batch[:, 0], batch[:, 1], bins=50, range=[(-5,5), (-5,5)], density=True)
    #     pl.xlim(-5, 5)
    #     pl.ylim(-5, 5)        
    #     pl.contour(X, Y, Z, levels=10, lw=2)
    #     pl.show()
```
